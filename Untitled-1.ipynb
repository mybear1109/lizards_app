```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import joblib
import os
```

```python
df = pd.read_csv('data/Lizards.csv', thousands=',')
```

```python
df
```

```python
# ✅ 기존 컬럼명 -> 변경할 컬럼명 매핑
df = df.rename(columns={
    "image_id": "Image",           # image_id -> Image
    "trans": "Transaction",        # trans -> Transaction
    "type": "Species",             # type -> Species
    "size": "Size",                # size -> Size
    "morph": "Morph",              # morph -> Morph
    "price": "Price",              # price -> Price
    "text": "Description",         # text -> Description
    "date": "Date"                 # date -> Date
})
```

```python
EXPECTED_COLUMNS=["Date", "Image", "Size", "Species", "Confidence", "Morph","Price"]
```

```python
missing_columns = [col for col in EXPECTED_COLUMNS if col not in df.columns]
for col in missing_columns:
    df[col] = None  # 기
```

```python
df = df[EXPECTED_COLUMNS]
```

```python
df.to_csv("data/Lizards.csv", index=False)
```
